```python
import pandas as pd
```

## Reading moon data from csv

```python
data = pd.read_csv("Moons_and_planets.csv",delimiter=',',header='infer')
data
```

```python
data.columns = ['Satellite','Planet','Diameter']
data.head()
```

## Counting number of moons of each planet

```python
Planets = ['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune','Pluto']
```

```python
print("Number of moons of each planet \n")

for planet in Planets:
    count = 0                               # counter for moons of particular planet
    for index, row in data.iterrows():      # Iterate over each row 
        if row['Planet'] == planet:
            count+=1                        
    print(planet,count)
```

## Ordering the moons according to size

```python
print("Moons in order of diameters with respective planets")
print("-"*30)
for index, row in data.sort_values(by='Diameter',ascending=False).iterrows():
    print("{:<15}{:<8}{:>5}".format(row['Satellite'],row['Planet'],row['Diameter']))
```
